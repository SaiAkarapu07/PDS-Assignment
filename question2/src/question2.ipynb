```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
```

```python
# Load the dataset
# Load the dataset from a CSV file
student = pd.read_csv("StudentsPerformance.csv")
student
```

```python
#Data processing

#Convert Frailty column to binary
label = LabelEncoder()
student['gender']=label.fit_transform(student['gender'])
student['race/ethnicity']=label.fit_transform(student['race/ethnicity'])
student['parental level of education']=label.fit_transform(student['parental level of education'])
student['lunch']=label.fit_transform(student['lunch'])
student['test preparation course']=label.fit_transform(student['test preparation course'])


student.to_csv('CleanedData.csv',index=False)
student
```

```python




# View the first 5 rows of the dataset
print(student.head())

# 1. Scatter plot of math score vs. reading score with color-coded gender
sns.scatterplot(x="math score", y="reading score", data=student, hue="gender")
plt.show()

# 2. Histogram of writing scores by gender
sns.histplot(x="writing score", data=student, hue="gender", multiple="stack")
plt.show()

# 3. Violin plot of race/ethnicity vs. math score
sns.violinplot(x="race/ethnicity", y="math score", data=student)
plt.show()

# 4. Correlation Plot / Heat map
sns.heatmap(student.corr(), cmap="coolwarm", annot=True)
plt.title("Correlation Matrix")
plt.show()

# 5. Stacked bar chart of lunch type vs. pass/fail status by race/ethnicity
sns.boxplot(data=student, x="race/ethnicity", y="math score")
plt.title("Math Score by Race/Ethnicity")
plt.xlabel("Race/Ethnicity")
plt.ylabel("Math Score")
plt.show()
```
