```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the data into a DataFrame
frality_df = pd.read_csv("raw data.csv")
frality_df
```

```python
# Remove any duplicates
frality_df.drop_duplicates(inplace=True)

# Convert the 'Frailty' column to boolean
frality_df['Frailty'] = frality_df['Frailty'].map({'Y': True, 'N': False})
```

```python
# Convert the data types for the other columns as needed
frality_df['Height (Inches)'] = frality_df['Height (Inches)'].astype(float)
frality_df['Weight (Pounds)'] = frality_df['Weight (Pounds)'].astype(float)
frality_df['Age'] = frality_df['Age'].astype(int)
frality_df['Grip strength'] = frality_df['Grip strength'].astype(int)

# Create a new column for BMI
frality_df['BMI'] = frality_df['Weight (Pounds)'] / (frality_df['Height (Inches)'] ** 2) * 703
frality_df
```

```python
# Save the cleaned data to a new file
frality_df.to_csv("cleaned_data.csv", index=False)
```

```python
# Compute pairwise correlations between the columns
print(frality_df.corr())
```

```python
# Create a histogram of the 'Age' column
sns.histplot(data=frality_df, x='Age')
plt.show()
```

```python
# Create a scatter plot of 'Weight' vs. 'Height'
sns.scatterplot(data=frality_df, x='Height (Inches)', y='Weight (Pounds)', hue='Frailty')
plt.show()
```
